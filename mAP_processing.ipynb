```python
import os
import numpy as np
import cv2
import yolo_modified as yolo
```

```python
gt_save_path = "mAP/input/ground-truth"
label_path = "model_data/MIR_test.txt"
label_dict = {0:'guitar'}
```

```python
with open(label_path) as f:
    for content in f.readlines():
        file_n = content.split(' ')[0]
        file_n = file_n.split('/')[2].split('.')[0]+'.txt'
        file = open(gt_save_path+"/"+file_n,'w')
        for box in content.split(' ')[1:]:
            box_splitted = box.split(',')
            row_i = label_dict[int(box_splitted[4])]
            for c in box_splitted[:4]:
                row_i = row_i+' '+c
            file.write(row_i)
            file.write('\n')
        file.close()
```

```python
model = yolo.YOLO()
```

```python
pr_save_path = "mAP/input/detection-results"
```

```python
with open(label_path) as f:
    for content in f.readlines():
        file_n = content.split(' ')[0]
        frame = cv2.imread(file_n)
        image = yolo.Image.fromarray(frame)
        out_boxes,scores, out_classes = model.detect_image_rectang(image)
        file_n = file_n.split('/')[2].split('.')[0]+'.txt'
        file = open(pr_save_path+"/"+file_n,'w')
        for i in range(0,len(out_boxes)):
            row_i = label_dict[int(out_classes[i])]+' '+str(scores[i])+' '+str(out_boxes[i][1])+' '+str(out_boxes[i][0])+' '+str(out_boxes[i][3])+' '+str(out_boxes[i][2])
            file.write(row_i)
            file.write('\n')
        file.close()
```
